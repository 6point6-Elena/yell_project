```python
import requests
import json

BASE_URL = 'https://api.yelp.com/v3'
API_KEY = 'ertgcPgirJwPK3bFbqyD9Binxc8wdgdEj-WPH3hWw2QARQte80IcA5k04gw4h_x19wKNxW78amCCz2ZLv7jAowrXkLxoX3gokqwqSA1oqvcbsbBpZnUxZu6sh4K1YnYx'
Location = 'London'

headers = {
    'Authorization': f'Bearer {API_KEY}'
}

response = requests.get(f"{BASE_URL}/businesses/search?location={Location}", headers=headers,auth=None)
print(response.status_code)
print(json.loads(response.content))
```

```python
obj = json.loads(response.content)
json_formatted_str = json.dumps(obj, indent=4)
print(json_formatted_str)
```

```python
import pandas as pd

business_id = []
business_name = []
business_rating = []
business_review_count = []

for business in obj['businesses']:
    business_id.append(business['id'])
    business_name.append(business['name'])
    business_rating.append(business['rating'])
    business_review_count.append(business['review_count'])
    

df = pd.DataFrame(list(zip(business_id, business_name, business_rating, business_review_count)), columns=['id', 'name', 'rating', 'review count'])

print(df)
                                                                          


```
